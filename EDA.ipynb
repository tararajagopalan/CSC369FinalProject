```python
#finding the most commonly used services

most_common_services_df_2022 = con.execute("""
SELECT
    HCPCS_Cd,
    HCPCS_Desc,
    Sum(Tot_Benes) as total_beneficiaries
FROM 'medicare_2022.parquet'
GROUP BY HCPCS_Cd, HCPCS_Desc
ORDER BY total_beneficiaries DESC
LIMIT 10
""").df()

most_common_services_df_2022
```
